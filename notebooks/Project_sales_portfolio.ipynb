```python
import  numpy as np
import pandas as pd
```

```python
# Read with latin-1 encoding (handles the problematic characters)
df = pd.read_csv('superstore.csv', encoding='latin-1')

# Save back as proper UTF-8
df.to_csv('superstore_utf8.csv', index=False, encoding='utf-8')
```

```python
# Remove spaces and make all column names lowercase
df.columns = df.columns.str.strip()           
df.columns = df.columns.str.lower()           
df.columns = df.columns.str.replace(' ', '_')
```

```python
print(df.columns)
```

```python
# Convert Order Date and Ship Date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.head(2)
```

```python
# Make sure Sales, Profit, Discount are numeric
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
print(df.dtypes)
```

```python
#handling missing values 
# Check missing values again after type conversion
print(df.isnull().sum())
```

```python
# Verify no missing values remain
print(df.isnull().sum())
```

```python
# For numeric columns â fill missing with 0 or median
df['sales'].fillna(df['sales'].median(), inplace=True)
df['profit'].fillna(0, inplace=True)
df['discount'].fillna(0, inplace=True)

# For text columns â fill with 'Unknown'
df['region'].fillna('Unknown', inplace=True)
df['customer_name'].fillna('Unknown', inplace=True)

# Verify no missing values remain
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
#check duplicates before  
print(df.duplicated().sum())

#remove duplicate rows 
df.drop_duplicates(inplace = True)
```

```python
#PROFIT MARGIN ADDING NEW COLUMN TO THE EXISTING DATA 
df['profit_margin_percentage'] = round((df['profit'] /df['sales']) * 100, 2)
print(df.columns)
print(df.profit_margin_percentage)
```

```python
#SHIPPING DAYS (HOW LONG DELIVERY TOOK)
df['shipping_days'] = (df['ship_date'] -df['order_date']).dt.days
```

```python
# 3. Extract Year, Month, Quarter from Order Date
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.strftime('%b')  # Jan, Feb...
df['order_quarter'] = df['order_date'].dt.quarter
df['month_year'] = df['order_date'].dt.strftime('%b-%Y')     # Jan-2021
```

```python
print(df[['order_date', 'order_year', 'order_month', 
          'order_quarter', 'shipping_days', 
          'profit_margin_percentage']].head())
```

```python
# Check for negative shipping days (data entry error)
print("Negative shipping days:", (df['shipping_days'] < 0).sum())

# Remove rows where shipping days is negative
df = df[df['shipping_days'] >= 0]

# Check for zero or negative sales (invalid orders)
print("Zero/Negative sales:", (df['sales'] <= 0).sum())

# Check profit margin extremes
print(df['profit_margin_percentage'].describe())

# See products with extreme negative margin
print(df[df['profit_margin_percentage'] < -50][['product_name', 
       'sales', 'profit', 'profit_margin_percentage']].head(10))
```

```python
# Final shape check
print("Final dataset shape:", df.shape)

# Final missing value check
print("Missing values:\n", df.isnull().sum())

# Preview cleaned data
print(df.head())

# Save cleaned file
df.to_csv('superstore_utf8.csv', index=False, encoding='utf-8')

print(" Data cleaning complete! File saved as superstore_utf8.csv")
```
